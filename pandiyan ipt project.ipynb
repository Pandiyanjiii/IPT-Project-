```python
import numpy as tr
import pandas as dc
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
dataset = sns.load_dataset('titanic')
```

```python
dataset
```

```python
dataset.info()
```

```python
data = dataset.drop(['survived','sex','age','sibsp','parch','embarked','class','who','adult_male','deck','embark_town','alive','alone'],axis=1)
```

```python
data
```

```python
x = data.iloc[:,-1].values.reshape(-1,1)
y = data.iloc[:,0].values.reshape(-1,1)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.title('pclass vs fare')
plt.xlabel('pclass')
plt.ylabel('fare')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
y_pred = lm.predict(x_test)
```

```python
y_pred
```

```python
check = dc.DataFrame(x_test,columns=['titanic'])
```

```python
check['Actual fare'] = y_test
```

```python
check['Predicted fare'] = y_pred
```

```python
check
```

```python
plt.scatter(x_test,y_test,color='pink')
plt.plot(x_test,y_pred,color='yellow')
plt.title('Linear Regression')
plt.xlabel('pclass')
plt.ylabel('fare')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
